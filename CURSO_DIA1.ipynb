## Operaciones Matematicas

```python
# 4to nivel de importancia
# Suma ordinaria
print("Suma ordinaria: ")
print(3 + 4) #un 7
variable1,variable2 = 20,40 #un 60
print(variable1 + variable2)

#Resta ordinaria
print("Resta ordinaria: ")
print(2 - 5) # -3

#Declaracion de variables
v1 = 10.4
v2 = 15
resultado = v2 - v1 # 4.6
print(resultado)
```

```python
# 3er nivel de importancia
#Multiplicacion ordinaria
print("Multipliacion ordinaria")
M1, M2 = 2,3.14
print(M1 * M2) #6.28
#Division ordinaria
print("-------------")
print("Division ordinaria")
M1,M2,M3,M4,M5,M6 = 2,5,6,5,4,5
print(M1 / M2)
print(M3 / M4)
print(M5 / M6)
#Division de Piso
print("-------------")
print("Division de Piso")
print(M1 // M2)
print(M3 // M4)
print(M5 // M6)
print( 6 / 2 *3)
```

```python
#Potencias
print("Potencias")
print(2 ** 3) #8
P1,P2 = 7,2
print(P1 ** P2) #49
```

```python
# Area de un circulo
pi = 3.14
r = 4
A = pi * (r ** 2)
A
```
